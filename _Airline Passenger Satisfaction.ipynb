```python
# Check for missing values
print(df.isnull().sum())
```

```python
from sklearn.impute import KNNImputer
import numpy as np
# Initialize the KNN imputer
imputer = KNNImputer(n_neighbors=5)

# Apply KNN imputation to 'Arrival Delay in Minutes'
imputed_values = imputer.fit_transform(df[['Arrival Delay in Minutes']])
# Update only the missing values in 'Arrival Delay in Minutes'
df['Arrival Delay in Minutes'] = np.where(df['Arrival Delay in Minutes'].isnull(),
                                          imputed_values[:, 0],
                                          df['Arrival Delay in Minutes'])
# Check if missing values are filled
print(df['Arrival Delay in Minutes'].isnull().sum())
```

```python
# Check for outliers
import pandas as pd

# Function to detect outliers using the IQR method
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Detect outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Check for outliers in relevant numerical columns
numerical_columns = df.select_dtypes(include=['int64', 'float64']).columns
for column in numerical_columns:
    outliers = detect_outliers_iqr(df, column)
    if not outliers.empty:
        print(f"Outliers detected in {column}: {outliers.shape[0]} rows")
    else:
        print(f"No outliers detected in {column}")
```

```python
# Remove the outliers
# Function to remove outliers using the IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile Range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Remove outliers
    df_filtered = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    
    return df_filtered

columns_with_outliers = ['Flight Distance', 'Checkin service', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']

for column in columns_with_outliers:
    df = remove_outliers_iqr(df, column)
print(f"Shape after removing outliers: {df.shape}")
```

```python
from sklearn.preprocessing import LabelEncoder
df_encoded = df.copy()
le = LabelEncoder()
categorical_features = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']
for feature in categorical_features:
    df_encoded[feature] = le.fit_transform(df_encoded[feature])
corr_matrix = df_encoded.corr()
plt.figure(figsize=(25, 20))
sns.heatmap(corr_matrix, annot=True, cmap='Blues', fmt='.2f', linewidths=.5)
plt.title('Correlation Heatmap with Categorical Variables')
plt.show()
```

```python
abs_corr_matrix = corr_matrix.abs()
# Find pairs with high correlation
threshold = 0.5
high_corr_pairs = (abs_corr_matrix.where(abs_corr_matrix > threshold)
                   .stack()
                   .reset_index()
                   .rename(columns={0: 'Correlation'}))

# Remove self-correlations
high_corr_pairs = high_corr_pairs[high_corr_pairs['level_0'] != high_corr_pairs['level_1']]
# Remove duplicate pairs
high_corr_pairs = high_corr_pairs.drop_duplicates(subset=['Correlation'])
# Sort by correlation value
high_corr_pairs = high_corr_pairs.sort_values(by='Correlation', ascending=False)
print(high_corr_pairs)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
# Define the models
models = {
    'GaussianNB': GaussianNB(),
    'GradientBoostingClassifier': GradientBoostingClassifier(),
    'LinearSVC': LinearSVC(max_iter=10000),
    'LogisticRegression': LogisticRegression(max_iter=10000),
    'MLPClassifier': MLPClassifier(max_iter=10000),
    'RandomForestClassifier': RandomForestClassifier()
}
X = df_encoded[features_of_interest]
y = df_encoded['satisfaction']

# Perform k-fold cross-validation
results = {}
for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')  # 5-fold cross-validation
    results[name] = scores.mean()
for model_name, mean_score in results.items():
    print(f'{model_name}: Mean Accuracy = {mean_score:.4f}')
```
