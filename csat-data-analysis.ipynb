
## ✈️ Passenger Satisfaction Insights (Based on Correlation Patterns)

1. **Customer experience outweighs operational issues**

   * Factors like **online boarding, inflight entertainment, seat comfort, legroom, and cleanliness** show stronger correlations (0.3–0.5) with satisfaction than delays do (\~0.2).
   * This means passengers are more forgiving of delays if the **onboard experience** is good.

2. **Online boarding is a game-changer**

   * With the highest correlation (\~0.50), **ease of boarding** is the single strongest predictor of satisfaction.
   * Fast, seamless boarding sets the tone for the entire journey.

3. **Entertainment, comfort, and cleanliness form the “experience cluster”**

   * Passengers who enjoy entertainment, comfort, and a clean cabin tend to rate overall satisfaction much higher.
   * These features are interlinked — improving one (e.g., seat comfort) positively influences perceptions of others.

4. **Inflight service strongly ties to operational trust**

   * **Service quality correlates with baggage handling and on-board service**.
   * Passengers view staff attentiveness and smooth logistics (baggage, check-in) as part of the same service quality package.

5. **Delays matter, but not the most**

   * **Arrival and departure delays are almost redundant (0.96 correlation)**, meaning one explains the other.
   * While delays hurt satisfaction, they are **secondary to direct comfort and service experiences**.
   * Passengers are more accepting of operational delays if inflight service compensates.

6. **Wifi and booking convenience overlap**

   * Ratings for **inflight wifi** and **ease of booking** are strongly correlated (\~0.7).
   * This suggests that **digitally savvy passengers** value both. Investing in a seamless **digital journey** (pre-flight + inflight) captures this segment.

7. **Two main satisfaction drivers emerge**

   * **Experience cluster:** boarding, comfort, entertainment, cleanliness, legroom.
   * **Operational cluster:** delays, baggage handling, check-in, wifi.
   * The **experience cluster dominates satisfaction** (direct link), while operational cluster supports indirectly.

8. **Passenger psychology = perception-driven**

   * Satisfaction is not just about *objective performance* (like delay minutes).
   * It’s about **how passengers feel**: if the cabin is clean, staff attentive, and the process smooth, satisfaction remains high even with moderate delays.

---

✅ **Strategic takeaway for airlines**:
To maximize passenger satisfaction and loyalty, focus resources on **improving inflight experience (boarding, entertainment, comfort, cleanliness)** rather than only operational efficiency. Delays should be minimized, but passengers forgive them if the **journey feels pleasant**.

# curve fitting process

Symbolic Regression with splines to derive business model equation

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import SplineTransformer
from sklearn.metrics import roc_auc_score, accuracy_score
from pysr import PySRRegressor 
traindf['satisfaction_num'] = traindf['satisfaction'].map(
    {'neutral or dissatisfied': 0, 'satisfied': 1}
)
y = traindf['satisfaction_num'].values 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

feature_cols = [
    'Online boarding',
    'Inflight wifi service',
    'Seat comfort',
    'Inflight entertainment',
    'Inflight service',
    'Leg room service',
    'Cleanliness',
    'Checkin service',
    'Arrival Delay in Minutes',
    'Departure Delay in Minutes'
]
X = traindf[feature_cols]
# Scale features for regression stability
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
    
) 
model = PySRRegressor(
    niterations=20,                # number of iterations 
    binary_operators=["+", "-", "*", "/", "pow"],
    unary_operators=["exp", "log", "sqrt"],
    loss="loss(x, y) = (x - y)^2",   # MSE
    model_selection="best",          # best equation by validation loss
)

model.fit(X_train, y_train)
print(model)
```

```python
from scipy.optimize import curve_fit
import numpy as np

# Define a flexible non-linear model
def revenue_model(x, a, b, c, d, e, f):
    # Feature mapping:
    # x[:,0] = Seat comfort
    # x[:,1] = Inflight entertainment
    # x[:,2] = Inflight service
    # x[:,3] = Leg room service
    # x[:,4] = Arrival Delay in Minutes
    # x[:,5] = Departure Delay in Minutes
    return (
        a * np.sqrt(x[:,0]) + 
        b * np.log1p(x[:,1]) + 
        c * np.sqrt(x[:,2]) + 
        d * np.sqrt(x[:,3]) - 
        e * np.log1p(x[:,4] + 1) -   # negative effect for delays
        f * np.log1p(x[:,5] + 1)     # negative effect for delays
    )

# Prepare feature matrix X and target y
feature_cols = [
    'Seat comfort',
    'Inflight entertainment',
    'Inflight service',
    'Leg room service',
    'Arrival Delay in Minutes',
    'Departure Delay in Minutes'
]
X = traindf[feature_cols].values
y = traindf['satisfaction_num'].values

# Fit the curve
params, cov = curve_fit(revenue_model, X, y)
# Print optimized coefficients
param_names = ['a', 'b', 'c', 'd', 'e', 'f']
for name, value in zip(param_names, params):
    print(f"{name} = {value:.4f}")
# Predict satisfaction using the fitted model
y_pred = revenue_model(X, *params)
```

```python
X = df[['Passengers_per_day', 'Average_Fare', 'Service_Index', 'Ancillary_per_passenger']].values
y = df['Revenue'].values

# Standardize features (mean=0, std=1)
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1,1)).ravel()
```

```python
# derive revenue optimization eqs using genetic algorithm
model = PySRRegressor(
    niterations=50,
    binary_operators=["+", "-", "*", "/"],
    unary_operators=["sin", "cos", "exp", "log"],
    population_size=100,
    model_selection="best",
    complexity_of_operators=None,  # DO NOT set as integer
    constraints=None
)

model.fit(X_scaled, y_scaled)
```

```python
# Print the best equation (lowest loss)
best_eq = model.get_best()
print(best_eq)
```
